```python
#aval libarary ro import kardim, bad maghadire x va y ro dadim, rang, title va name mehvar X va Y 
#ro moshakhas kardim
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[4,5,4,6,3,2,1]
plt.plot(x,y,color="green")
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#agar inja "o" ro bezarim, bejaye khat, noghtee be ma neshan midahad.
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[4,5,4,6,3,2,1]
plt.plot(x,y,"o",color="green")
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#hala age bekhaym bejaye "o" , "*" bezarim va hamchenin in noghat ro be ham ba khat vasl konim
#be soorate zir amal konim
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[4,5,4,6,3,2,1]
plt.plot(x,y,marker="*",color="green")
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#baraye rasme bar graph ham be sorate zir amal mikonim
x=["red","green","blue","yellow"]
y=[4,5,4,6]
plt.bar(x,y,color="green")
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#baraye rasme pie chart ham be sorate zir amal mikonim
x=["red","green","blue","yellow"]
y=[4,11,4,6]
plt.pie(y,labels=x,autopct="%1.1f%%",colors=["red","green","blue","yellow"])
plt.show()
```

```python
#dat histogram zir, miad tedade random adaad beyne 
import numpy as np
x=np.random.rand(1000,1)
plt.hist(x,bins=8)
plt.show()
```

```python
#agar bekhahim heat map ijad konim az tabe imshow() estefade mikonim.
x=np.random.rand(64,64) # an array with 2 dimentions: 64 rows and 64 columns

plt.imshow(x,cmap="hot") #cmap range heatmap maro moshkhas mikonad ke inja be range zardo inast
plt.colorbar() # inja baze rangharo mashaskhas kardim ke dar samte rast ghabele moshahede ast
plt.show()
```
