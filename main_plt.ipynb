```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Given points 
x = np.array([230.1, 44.5, 17.2, 151.5,180.8])
y = np.array([22.1,10.4,9.3,18.5,12.9])

# Regression line
m = 0.6
b = 1
x_line = np.linspace(0, 5, 100)
y_line = m * x_line + b

# Plotting
plt.scatter(x, y, color='red', label='Data points')
plt.plot(x_line, y_line, color='blue', label='Regression line: y = 0.6x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Regression Line')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Create the arrays for x and y
x = np.array([230.1, 44.5, 17.2, 151.5, 180.8]).reshape(-1, 1)
y = np.array([22.1, 10.4, 9.3, 18.5, 12.9])

# Fit a linear regression model
model = LinearRegression()
model.fit(x, y)

# Predict the y values using the model
y_pred = model.predict(x)

# Calculate the mean squared error
mse = mean_squared_error(y, y_pred)

# Print the results
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
print(f"Mean Squared Error: {mse}")

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, y_pred, color='red', label='Regression Line')

# Plot the error lines
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y[i], y_pred[i]], 'k--')

# Annotate the plot with MSE
plt.text(min(x), max(y), f'MSE: {mse:.2f}', fontsize=12, color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression with Mean Squared Error')
plt.legend()
plt.grid(True)
plt.show()

```
