```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

#data generation
y=x^2+5*x+3

```python
x=np.arange(-80,81,0.5) # range adadi ke mikham ro moshakhas mikonam ke beyne -80 ta 80 hast ba tole gaame 0.5
y=x**2+5*x+3
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) # jodasazi dade train va dade test khodemon
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

#model defintion 1

```python
model_1=keras.Sequential()
model_1.add(keras.layers.Dense(1024,activation='relu',input_shape=[1]))#tabe add moshakhas mikone ke che laye mikhahim be model ezafe konim
#ma dar khate bala, darim layeye bad az laye vorodi ro tarif mikonim, dar inja darim migim in avalin hidden layer ma hast, 1024 neuron dare
#yedone vorodi migire va activation function ham tabe relu hast
#dar tensorflow niazi nist ke laye vorodi ro tarif konim va inja ba tarif laye hidden avali, laye vorodi ham khodesh tanzim mishe
#
model_1.add(keras.layers.Dense(units=1))

```

```python
model_1.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model_1.summary()
```

```python
hist_1=model_1.fit(x_train,y_train,batch_size=64, epochs=500,validation_data=(x_test,y_test)) #dade haye x_train va y_train man ro bardar
#be batch-size haye 64 ee tabdil kon va har epoch ro 50 begzar, validation_data ham az x_test va y_test bardar
```

#model definition 2

```python
model_2=keras.Sequential()# dar model dovomemon bejaye inke laye aval 1024 ta layer dashte bashe, ma 4 ta laye 256 tarif mikonim
#ba in kar mikhaym bebin farghe ye laye 1024 neruni ba 4 ta laye 256 neruni che tasiri dar model ma baraye train migzare
model_2.add(keras.layers.Dense(units=256,activation='relu',input_shape=[1]))
model_2.add(keras.layers.Dense(units=256,activation='relu'))
model_2.add(keras.layers.Dense(units=256,activation='relu'))
model_2.add(keras.layers.Dense(units=256,activation='relu'))
model_2.add(keras.layers.Dense(units=1))
```

```python
model_2.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model_2.summary()
```

```python
hist_2=model_2.fit(x_train,y_train,batch_size=64, epochs=500,validation_data=(x_test,y_test))
```

```python
y_pred1=model_1.predict(x_test)
y_pred2=model_2.predict(x_test)
```

```python
plt.scatter(x_test,y_test,color='red',linewidths=3)
plt.scatter(x_test,y_pred1,color='blue')
plt.scatter(x_test,y_pred2,color='green')
```

```python
plt.plot(hist_1.history['loss'],linewidth=5,color='orange')
plt.plot(hist_1.history['val_loss'],color='blue')
plt.plot(hist_2.history['loss'],linewidth=5,color='green')
plt.plot(hist_2.history['val_loss'],color='red')
plt.show()
```
